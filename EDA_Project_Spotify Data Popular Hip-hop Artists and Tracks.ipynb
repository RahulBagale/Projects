6.Among the top 5 most popular artists, which artist has the highest popularity on average?

7.For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist


```python
df2 = df1[['Artist','Track Name','Popularity']]
```

```python
# Finding the top 5 Artist by comparing their popularity mean
df4 = df2.groupby('Artist')['Popularity'].mean().sort_values(ascending = False).head(5)
df4 = pd.DataFrame(df4)
l1  = list(df4.index)
```

```python
# Loop for printing the top 5 most popular artist and top 5 most popular tracks
for i in l1:
  df10 = df2[df2['Artist']==i]
  print('-'*30)
  print(df10.sort_values(by='Popularity',ascending = False).head(5))
```
